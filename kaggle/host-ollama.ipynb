```python
import subprocess
import time

def run_command(command):
    process = subprocess.Popen(command,
                               shell=True,
                               stdout=subprocess.PIPE,
                               stderr=subprocess.PIPE)
    return process

process = run_command("killall -9 ollama")
process = run_command("nohup ollama serve&")

print("Process ID:", process.pid)
```
